# Work with other infrastuctures besides streets

Use OSMnx to download and visualize a power line network and a subway system.

  - [Overview of OSMnx](http://geoffboeing.com/2016/11/osmnx-python-street-networks/)
  - [GitHub repo](https://github.com/gboeing/osmnx)
  - [Examples, demos, tutorials](https://github.com/gboeing/osmnx-examples)
  - [Documentation](https://osmnx.readthedocs.io/en/stable/)
  - [Journal article/citation](http://geoffboeing.com/publications/osmnx-complex-street-networks/)

```python
import osmnx as ox
ox.config(use_cache=True, log_console=True)
```

```python
point = (33.9017,-116.9594)
dist = 100000
north, south, east, west = ox.bbox_from_point(point, distance=dist)
```

```python
point = (55.8519813,-4.2582198)
dist = 10000
north, south, east, west = ox.bbox_from_point(point, distance=dist)

cf = """
     ["ref"!~""]
     ["area"!~"yes"]
     ["highway"!~"motorway|primary|secondary"]
     """

G = ox.graph_from_bbox(north=north, south=south, east=east, west=west,
                       retain_all=True, truncate_by_edge=True, simplify=True,
                       network_type='none', infrastructure='way["tunnel"]', custom_filter=cf)
fig, ax = ox.plot_graph(ox.project_graph(G),fig_height=10, fig_width=15)
```

```python
G = ox.graph_from_bbox(north=north, south=south, east=east, west=west,
                       retain_all=True, truncate_by_edge=True, simplify=True,
                       network_type='none', infrastructure='way["railway"]', custom_filter=cf)
fig, ax = ox.plot_graph(ox.project_graph(G),fig_height=10, fig_width=15)
```
